## 4. Variables

#### Python Variables

- Variables are containers for storing data.
- A variable is created at the moment you assign a value to it.
- Variables do not need to be declared with any particular data type they can even change type after they have been set.

```python
a = 12
b = 'Hamza'
```

```python
print(a)
print(b)
```

a & b are variables having value 12 assigned in variable a & 'Hamza' assigned in variable b

#### Variable Names 

- A variable can have a short name (like a and b) or a more descriptive name (age, firstname, total_price). 
- Rules for Python variables:
    - A variable name must start with a letter or the underscore character
    - A variable name cannot start with a number
    - A variable name can only contain alpha-numeric characters and underscores (A-z, 0-9, and _ )
    - Variable names are case-sensitive (car, Car and CAR are three different variables)

#### Assign Multiple Values

```python
name, age, gender = 'Hamza', 21, 'M'
```

```python
print(name)
print(age)
print(gender)
```

```python
word = s_word = Tword = 'Hello'
```

```python
print(word)
print(s_word)
print(Tword)
```

You can also use a variable as an Output variable using concatenation denoted by + or ,

```python
print('My Name Is: ' + name) #name variable assigned above
```

```python
print('My Name Is: ', name) #name variable assigned above
```

```python
print('My Name Is: ', name, 'And Age: ', age)
```

```python
print(name+gender) # Two String Variables
```

```python
#Error will occur if we used + to concatenate a String & an Integer 
print(name+age)
```

```python
#But Error will be resolved if we used , to concatenate a String & an Integer 
print(name,age)
```
